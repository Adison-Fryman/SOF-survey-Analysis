```python
#Both .at and .loc are used to access specific elements in a pandas DataFrame or Series, but they have different use cases.

#.at is used to access a single scalar value based on the row and column labels. It is faster than .loc because it accesses only one element at a time. For example, to access the value in the first row and second column of a DataFrame, you can use .at like this:


#df.at[0, 'column2']
#.loc, on the other hand, is used to access multiple values or a subset of a DataFrame or Series based on row and column labels.
#It can access multiple elements at once, which makes it slower than .at. 
#For example, to access all rows and columns of a DataFrame that satisfy a certain condition, you can use .loc like this:

#df.loc[df['column1'] == 'value1', ['column2', 'column3']]
#In summary, if you need to access a single scalar value in a DataFrame or Series, use .at. 
#If you need to access multiple values or a subset of a DataFrame or Series, use .loc.
```

```python
# a view and a copy refer to different ways of referencing a NumPy array or a Pandas DataFrame. 
#Understanding the difference between a view and a copy is important because modifying one may or may not affect the other.

#A VIEW is a reference to the original array or dataframe. When you create a view, it does not create a new object in memory. 
#Instead, it points to the same memory location as the original array or dataframe. 
#This means that any changes made to the view will affect the original array or dataframe.

#A COPY, on the other hand, creates a new object in memory. Any changes made to the copy will not affect the original array 
#or dataframe.

#To create a view in NumPy, you can use the slice notation with the colon operator. 
#In Pandas, you can use the .loc or .iloc methods to create a view of a DataFrame.

#To create a copy in NumPy, you can use the .copy() method. 
#In Pandas, you can use the .copy() method to create a copy of a DataFrame.

#It is important to understand the difference between a view and a copy because modifying one may or may not affect the other. If you are unsure which one you are working with, you can use the .base attribute in NumPy or the .is_view() method in Pandas to check whether an array or dataframe is a view or a copy.
```

```python
#how to change all rows 4 kinds of ways
#apply (calling function on values) on a series 
#map (only works on a series, subs in new values in a series) 
#applymap applys a function to every value in a dataframe, It does not have the capability to filter rows based on a condition.
#replace- dont need to do EVERY value

#All these methods are used to modify the data in a DataFrame or Series. 
#The key difference between these methods is in the type of input they take and the way they apply the modifications to the data. 
#apply() and applymap() take a function as input, 
#while map() and replace() can take a function or a dictionary as input.

 #apply() and applymap() usually return a new DataFrame, which can be a view or a copy depending on the input and the operation performed. 
    #On the other hand, map() and replace() typically return a new Series, which is always a view.
    #In general, if the input data is a view of a larger DataFrame or Series, then the returned object is also a view. 
    #However, if the input data is a copy, then the returned object is also a copy.
```
