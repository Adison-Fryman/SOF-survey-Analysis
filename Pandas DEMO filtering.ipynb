```python
filt = df['Employment'] == 'Employed, full-time' # filtering column for a specific value, set as a varible, then print
```

```python
df[filt]
```

```python
df.loc[filt,'RemoteWork'] #second option filters data out but lets you grab other collomns
```

```python
filt2 = (df['Employment'] =='Employed, full-time') & (df['RemoteWork'] =='Fully remote') # option for filterig, good for looking at several columns and specific values, not so good for wanting multiple answers
```

```python
df.loc[filt2] # use .loc[filter variable] to print out results
```

```python
filt3 = (df['Employment'] =='Employed, full-time') | (df['RemoteWork'] =='Fully remote') # | this is the "or" opperator
```

```python
df.loc[filt3]
```

```python
high_salary = (df['CompTotal']>90000)
df.loc[high_salary,['Country','LanguageHaveWorkedWith','CompTotal' ]].head() # finding the high salary entries, and returning specific columns with it
# then using.head to preview the table
```

```python
low_salary = ~(df['CompTotal']>90000) # "~" is the not opperator
df.loc[low_salary].head()
```
