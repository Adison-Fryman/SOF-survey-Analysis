```python
import numpy as np

# define dictionary of currency converters
currencies_converters = {'EUR': .93, 'USD': 1, 'INR': 82.31, 'GBP': .82,'CAD': 1.37, 'BRL': 5.25, 'PLN': 4.39, 'AUD': 1.5}
# taken from goodle converter 3/14/2023

#triming unneeded characters from 'Currency'
df['Currency'] = df['Currency'].str[:3]

# function to convert currency
def convert_currency(currency):
    if currency in currencies_converters:
        return currencies_converters[currency]
    else:
        return np.NaN  # using np.NaN instead of NaN

# add new 'CurrencyConvert' column to DataFrame using apply() method
df['CurrencyConvert'] = df['Currency'].apply(convert_currency)


def get_interval(interval):
    if pd.isna(interval):
        return np.NaN
    elif interval == "Yearly":
        return 1
    elif interval == "Monthly":
        return 12
    elif interval == "Weekly":
        return 52
    else:
        raise ValueError("Invalid interval value: {}".format(interval))
        
        
        

df['TotConvertCOMP'] = df['CurrencyConvert'] * df['CompTotal'] * df['CompFreq'].apply(get_interval)

# look at what I have done with all the Nan take out, removing rows with NaN values using .dropna
# display updated DataFrame
df = df.dropna(subset=['CurrencyConvert', 'CompTotal', 'CompFreq']) # is how = 'any' the default?
#df = df.dropna(axis=1).....This will remove all columns that contain at least one NaN value.
#remove the columns by label/name
df = df.drop(columns=['LearnCodeOnline', 'LearnCodeCoursesCert'])
#remove the columns by slice
df = df.drop(columns=df.columns[df.columns.get_loc('OrgSize'):df.columns.get_loc('BuyNewTool')])
df
```
