```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents,country_uses_python],axis = 'columns',sort = True)
python_df
```

```python
python_df.rename(columns = {'Country':"numRespons", "LanguageHaveWorkedWith":'NumKnowPy'}, inplace = True) # renaming table lables
python_df
```

```python
python_df['PercentknowPy'] = ((python_df['NumKnowPy']/python_df["numRespons"]))*100
```

```python
python_df.sort_values(by ='PercentknowPy', ascending = True, inplace = True)#
#python_df = python_df.sort_values(by ='PercentknowPy', ascending = False).head(50)
python_df.head(50)
#python_df.loc['China']
```
